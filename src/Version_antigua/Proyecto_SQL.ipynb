## Creación de tablas

#### Campus

```python
# Primero eliminamos la tabla si ya existe (útil para reiniciar desde cero)
# DROP TABLE IF EXISTS es una operación segura que no da error si la tabla no existe
cursor.execute('DROP TABLE IF EXISTS Campus')

# Creamos la tabla Campus con SQL usando cursor.execute()
# Tipos de datos en SQLite:
#   - INTEGER: números enteros
#   - TEXT: cadenas de texto
#   - REAL: números decimales
#   - DATE: fechas
# Restricciones:
#   - PRIMARY KEY: identifica de forma única cada registro
#   - AUTOINCREMENT: genera automáticamente el ID (1, 2, 3, ...)
#   - NOT NULL: el campo no puede estar vacío (es obligatorio)
#   - UNIQUE: no puede haber valores duplicados (ej: dos emails iguales)
cursor.execute('''
    CREATE TABLE Campus (
        id_campus INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre_campus VCHAR(50) NOT NULL
    )
''')

# Guardamos los cambios en la base de datos
# commit() hace permanentes los cambios - sin esto, los cambios se pierden
conexion.commit()

print("✓ Tabla Campus creada exitosamente")
```

```python
# Primero eliminamos la tabla si ya existe (útil para reiniciar desde cero)
# DROP TABLE IF EXISTS es una operación segura que no da error si la tabla no existe
cursor.execute('DROP TABLE IF EXISTS Profesor')

# Creamos la tabla Profesor con SQL usando cursor.execute()
# Tipos de datos en SQLite:
#   - INTEGER: números enteros
#   - TEXT: cadenas de texto
#   - REAL: números decimales
#   - DATE: fechas
# Restricciones:
#   - PRIMARY KEY: identifica de forma única cada registro
#   - AUTOINCREMENT: genera automáticamente el ID (1, 2, 3, ...)
#   - NOT NULL: el campo no puede estar vacío (es obligatorio)
#   - UNIQUE: no puede haber valores duplicados (ej: dos emails iguales)
cursor.execute('''
    CREATE TABLE Profesor (
        id_profesor INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre_profesor VCHAR(100) NOT NULL,
        rol VCHAR(20) NOT NULL
    )
''')

# Guardamos los cambios en la base de datos
# commit() hace permanentes los cambios - sin esto, los cambios se pierden
conexion.commit()

print("✓ Tabla Profesor creada exitosamente")
```

#### Vertical

```python
# Primero eliminamos la tabla si ya existe (útil para reiniciar desde cero)
# DROP TABLE IF EXISTS es una operación segura que no da error si la tabla no existe
cursor.execute('DROP TABLE IF EXISTS Vertical')

# Creamos la tabla Vertical con SQL usando cursor.execute()
# Tipos de datos en SQLite:
#   - INTEGER: números enteros
#   - TEXT: cadenas de texto
#   - REAL: números decimales
#   - DATE: fechas
# Restricciones:
#   - PRIMARY KEY: identifica de forma única cada registro
#   - AUTOINCREMENT: genera automáticamente el ID (1, 2, 3, ...)
#   - NOT NULL: el campo no puede estar vacío (es obligatorio)
#   - UNIQUE: no puede haber valores duplicados (ej: dos emails iguales)
cursor.execute('''
    CREATE TABLE Vertical (
        id_vertical INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre_vertical VCHAR(50) NOT NULL,
        id_proyectos INTEGER,
        FOREIGN KEY (id_proyectos) REFERENCES Proyecto(id_proyectos) ON DELETE CASCADE
    )
''')

# Guardamos los cambios en la base de datos
# commit() hace permanentes los cambios - sin esto, los cambios se pierden
conexion.commit()

print("✓ Tabla Vertical creada exitosamente")
```

```python
import sqlite3
import pandas as pd

df1 = df_total
df2= df_5

# Conectar a la base de datos
conn = sqlite3.connect('bootcamp.db')
cursor = conn.cursor()

# 1. CREAR TABLAS (en orden sin dependencias primero)

cursor.execute('''
CREATE TABLE IF NOT EXISTS Campus (
    id_campus INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_campus TEXT UNIQUE NOT NULL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Vertical (
    id_vertical INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_vertical TEXT UNIQUE NOT NULL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Profesor (
    id_profesor INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_profesor TEXT NOT NULL,
    rol TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Promocion (
    id_promocion INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_promocion TEXT NOT NULL,
    fecha_promocion DATE,
    modalidad TEXT,
    id_campus INTEGER,
    id_profesor INTEGER,
    id_vertical INTEGER,
    FOREIGN KEY (id_campus) REFERENCES Campus(id_campus),
    FOREIGN KEY (id_profesor) REFERENCES Profesor(id_profesor),
    FOREIGN KEY (id_vertical) REFERENCES Vertical(id_vertical),
    UNIQUE(nombre_promocion, fecha_promocion, id_campus, id_vertical)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Alumno (
    id_alumno INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_alumno TEXT NOT NULL,
    email_alumno TEXT UNIQUE,
    id_promocion INTEGER,
    FOREIGN KEY (id_promocion) REFERENCES Promocion(id_promocion)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Proyectos (
    id_proyectos INTEGER PRIMARY KEY AUTOINCREMENT,
    id_alumno INTEGER,
    Proyecto_HLF TEXT,
    Proyecto_EDA TEXT,
    Proyecto_BBDD TEXT,
    Proyecto_ML TEXT,
    Proyecto_Deployment TEXT,
    Proyecto_WebDev TEXT,
    Proyecto_FrontEnd TEXT,
    Proyecto_Backend TEXT,
    Proyecto_React TEXT,
    Proyecto_FullStack TEXT,
    FOREIGN KEY (id_alumno) REFERENCES Alumno(id_alumno)
)
''')

# 2. INSERTAR CAMPUS
campus_unicos = pd.concat([df1['Campus'], df2['Campus']]).dropna().unique()
for campus in campus_unicos:
    cursor.execute('INSERT OR IGNORE INTO Campus (nombre_campus) VALUES (?)', (campus,))

# 3. INSERTAR VERTICALES
cursor.execute('INSERT OR IGNORE INTO Vertical (nombre_vertical) VALUES (?)', ('DS',))
cursor.execute('INSERT OR IGNORE INTO Vertical (nombre_vertical) VALUES (?)', ('FS',))

# 4. INSERTAR PROFESORES
profesores = df2[['Nombre', 'Rol']].drop_duplicates()
for _, row in profesores.iterrows():
    cursor.execute('''
        INSERT OR IGNORE INTO Profesor (nombre_profesor, rol) 
        VALUES (?, ?)
    ''', (row['Nombre'], row['Rol']))

conn.commit()

# 5. FUNCIÓN PARA DETERMINAR VERTICAL DEL ALUMNO
def determinar_vertical_alumno(row):
    """
    Determina la vertical del alumno basándose en sus proyectos.
    DS: HLF, EDA, BBDD, ML, Deployment
    FS: WebDev, FrontEnd, Backend, React, FullStack
    """
    proyectos_ds = ['Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 
                    'Proyecto_ML', 'Proyecto_Deployment']
    proyectos_fs = ['Proyecto_WebDev', 'Proyecto_FrontEnd', 'Proyecto_Backend', 
                    'Proyecto_React', 'Proyecto_FullSatck']
    
    # Contar proyectos no vacíos de cada vertical
    count_ds = sum(1 for p in proyectos_ds if pd.notna(row[p]) and str(row[p]).strip() != '')
    count_fs = sum(1 for p in proyectos_fs if pd.notna(row[p]) and str(row[p]).strip() != '')
    
    # Si tiene más proyectos DS, es DS; si tiene más FS, es FS
    if count_ds > count_fs:
        return 'DS'
    elif count_fs > count_ds:
        return 'FS'
    else:
        # Si empate o no tiene proyectos, intentar determinar por otros medios
        # Por defecto DS si tiene al menos un proyecto DS
        if count_ds > 0:
            return 'DS'
        else:
            return 'FS'

# Añadir columna de vertical a df1
df1['Vertical_Alumno'] = df1.apply(determinar_vertical_alumno, axis=1)

# 6. INSERTAR PROMOCIONES
# Crear todas las combinaciones únicas de (nombre_promocion, fecha, campus, vertical)
promociones_necesarias = df1[['Promoción', 'Fecha_comienzo', 'Campus', 'Vertical_Alumno']].drop_duplicates()

for _, promo_row in promociones_necesarias.iterrows():
    # Obtener id_campus
    cursor.execute('SELECT id_campus FROM Campus WHERE nombre_campus = ?', (promo_row['Campus'],))
    result_campus = cursor.fetchone()
    
    # Obtener id_vertical
    cursor.execute('SELECT id_vertical FROM Vertical WHERE nombre_vertical = ?', (promo_row['Vertical_Alumno'],))
    result_vertical = cursor.fetchone()
    
    if result_campus and result_vertical:
        id_campus = result_campus[0]
        id_vertical = result_vertical[0]
        
        # Buscar profesor de esta vertical en df2
        profesor_df2 = df2[
            (df2['Promocion'] == promo_row['Promoción']) & 
            (df2['Campus'] == promo_row['Campus']) &
            (df2['Vertical'] == promo_row['Vertical_Alumno'])
        ]
        
        if not profesor_df2.empty:
            # Obtener modalidad del primer profesor encontrado
            modalidad = profesor_df2.iloc[0]['Modalidad']
            
            # Insertar una promoción por cada profesor de esta combinación
            for _, prof_row in profesor_df2.iterrows():
                cursor.execute('SELECT id_profesor FROM Profesor WHERE nombre_profesor = ?', (prof_row['Nombre'],))
                result_profesor = cursor.fetchone()
                
                if result_profesor:
                    id_profesor = result_profesor[0]
                    
                    cursor.execute('''
                        INSERT OR IGNORE INTO Promocion 
                        (nombre_promocion, fecha_promocion, modalidad, id_campus, id_profesor, id_vertical) 
                        VALUES (?, ?, ?, ?, ?, ?)
                    ''', (promo_row['Promoción'], promo_row['Fecha_comienzo'], modalidad, 
                          id_campus, id_profesor, id_vertical))

conn.commit()

# 7. INSERTAR ALUMNOS
for _, row in df1.iterrows():
    # Buscar la promoción que corresponde exactamente a este alumno
    cursor.execute('''
        SELECT p.id_promocion 
        FROM Promocion p
        JOIN Campus c ON p.id_campus = c.id_campus
        JOIN Vertical v ON p.id_vertical = v.id_vertical
        WHERE p.nombre_promocion = ? 
        AND p.fecha_promocion = ?
        AND c.nombre_campus = ? 
        AND v.nombre_vertical = ?
        LIMIT 1
    ''', (row['Promoción'], row['Fecha_comienzo'], row['Campus'], row['Vertical_Alumno']))
    
    result = cursor.fetchone()
    
    if result:
        id_promocion = result[0]
        cursor.execute('''
            INSERT OR IGNORE INTO Alumno (nombre_alumno, email_alumno, id_promocion) 
            VALUES (?, ?, ?)
        ''', (row['Nombre'], row['Email'], id_promocion))
    else:
        print(f"⚠️ No se encontró promoción para: {row['Nombre']} - {row['Promoción']} - {row['Campus']} - {row['Vertical_Alumno']}")

conn.commit()

# 8. INSERTAR PROYECTOS
for _, row in df1.iterrows():
    cursor.execute('SELECT id_alumno FROM Alumno WHERE email_alumno = ?', (row['Email'],))
    result = cursor.fetchone()
    
    if result:
        id_alumno = result[0]
        cursor.execute('''
            INSERT INTO Proyectos 
            (id_alumno, Proyecto_HLF, Proyecto_EDA, Proyecto_BBDD, Proyecto_ML, 
             Proyecto_Deployment, Proyecto_WebDev, Proyecto_FrontEnd, Proyecto_Backend, 
             Proyecto_React, Proyecto_FullStack) 
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', (id_alumno, row['Proyecto_HLF'], row['Proyecto_EDA'], row['Proyecto_BBDD'], 
              row['Proyecto_ML'], row['Proyecto_Deployment'], row['Proyecto_WebDev'], 
              row['Proyecto_FrontEnd'], row['Proyecto_Backend'], row['Proyecto_React'], 
              row['Proyecto_FullSatck']))

conn.commit()
conn.close()

print("¡Base de datos creada exitosamente!")

# VERIFICACIÓN DETALLADA
conn = sqlite3.connect('bootcamp.db')
cursor = conn.cursor()

tablas = ['Campus', 'Vertical', 'Profesor', 'Promocion', 'Alumno', 'Proyectos']
print("\n=== RESUMEN DE REGISTROS ===")
for tabla in tablas:
    cursor.execute(f'SELECT COUNT(*) FROM {tabla}')
    print(f"{tabla}: {cursor.fetchone()[0]} registros")

print("\n=== DETALLE PROMOCIONES ===")
cursor.execute('''
    SELECT p.nombre_promocion, p.fecha_promocion, c.nombre_campus, 
           v.nombre_vertical, pr.nombre_profesor, p.modalidad
    FROM Promocion p
    JOIN Campus c ON p.id_campus = c.id_campus
    JOIN Vertical v ON p.id_vertical = v.id_vertical
    JOIN Profesor pr ON p.id_profesor = pr.id_profesor
    ORDER BY p.nombre_promocion, v.nombre_vertical
''')
print("Promoción | Fecha | Campus | Vertical | Profesor | Modalidad")
print("-" * 80)
for row in cursor.fetchall():
    print(f"{row[0]} | {row[1]} | {row[2]} | {row[3]} | {row[4]} | {row[5]}")

print("\n=== DISTRIBUCIÓN DE ALUMNOS POR VERTICAL ===")
cursor.execute('''
    SELECT v.nombre_vertical, COUNT(a.id_alumno) as num_alumnos
    FROM Alumno a
    JOIN Promocion p ON a.id_promocion = p.id_promocion
    JOIN Vertical v ON p.id_vertical = v.id_vertical
    GROUP BY v.nombre_vertical
''')
for row in cursor.fetchall():
    print(f"  {row[0]}: {row[1]} alumnos")

conn.close()
```
